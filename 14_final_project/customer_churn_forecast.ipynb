# Описание проекта

По заданию от компании «Ниединогоразрыва.ком» требуется произвести исследование и создать модель, прогнозирующую уход клиента. В качестве исходной информации получены исторические данные компании.
Исторические данные содержат:
- информацию о договорах
- персональные данные клиентов
- информацию об используемых услугах

Результаты проведенной работы планируется использовать для удержания текущих клиентов.

**Общие сведения о Заказчике.**

Заказчик осуществляет предоставление услуг проводной связи и сопутствующих дополнительных сервисов.

Основные типы предоставляемых услуг:

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL) или оптическую линию (Fiber optic).

Сопутствующие сервисы, предоставляемые оператором:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*);
- Возможность использования многоканальной линии(*MultipleLines*).

Клиентам доступны три способа начисления абонентской платы: ежемесячный, ежегодный и раз в два года.

**Предоставленные Заказчиком данные.**

Данные состоят из файлов, полученных из разных источников:
- `contract_new.csv` — информация о договоре;
- `personal_new.csv` — персональные данные клиента;
- `internet_new.csv` — информация об интернет-услугах;
- `phone_new.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.
Информация о договорах актуальна на 1 февраля 2020.

Для расчета продолжительности действующих договоров воспользуемся условно текущей датой( для данного исследования возьмем 1 февраля 2020 года).


```python
phone_data['phone'] = 1
```

```python
internet_data['internet'] = 1
```

```python
def exet(x):
    if x == 'No':
        return 0
    else:
        return 1
```

```python
contract_data['exited'] = contract_data['EndDate'].apply(exet)
```

```python
contract_data['duration'] = pd.to_numeric((
    pd.to_datetime((contract_data['EndDate'].replace('No', '2020-02-01')),
                   format='%Y-%m-%d') -
    pd.to_datetime(contract_data['BeginDate'], format='%Y-%m-%d'))\
    .apply(lambda x: x.days),
    downcast="integer").astype(int)
```

Теперь посмотрим, как распределяются значения в разных столбцах.
Для начала рассмотрим, как распределяются абоненты в категории расторжение договора, а затем рассмотрим как делятся значения других столбцов в зависимости от статуса договора(действующий/расторгнутый).

```python
data_all.groupby('exited')['exited'].hist()
plt.xlabel('статус договора')
plt.ylabel('договоров')
plt.show()
```

На гистограмме видно, что расторгнутых договоров значительно меньше, чем действующих. Из этого следует, что в данных есть дисбаланс классов.

Теперь посмотрим, как распределяются значения в других столбцах в зависимости от статуса договора(действующий/расторгнут) и используемых услуг.

```python
params_col = ['Type', 'PaperlessBilling', 'PaymentMethod',
              'gender', 'SeniorCitizen', 'Partner',
              'Dependents', 'phone','internet']
params_col_i = ['InternetService', 'OnlineSecurity', 'OnlineBackup',
              'DeviceProtection', 'TechSupport', 'StreamingTV',
              'StreamingMovies']
params_col_t = ['MultipleLines']
for param in params_col:
    data_all.groupby('exited')[param].hist(alpha=0.35)
    plt.legend(['действующий', 'расторгнутый'])
    plt.xlabel(param)
    plt.ylabel('договоров')
    plt.show()

for param in params_col_i:
    data_all[data_all['internet']==1].groupby(['exited'])[param]\
                                     .hist(alpha=0.35)
    plt.legend(['действующий', 'расторгнутый'])
    plt.xlabel(param)
    plt.ylabel('договоров')
    plt.show()

for param in params_col_t:
    data_all[data_all['phone']==1].groupby(['exited'])[param]\
                                  .hist(alpha=0.35)
    plt.legend(['действующий', 'расторгнутый'])
    plt.xlabel(param)
    plt.ylabel('договоров')
    plt.show()
```

По графикам видно, что количество расторгнутых договоров примерно одинаковое среди всех типов начисления оплаты и способов оплаты.

В категориях тип квитанции, пол, статус пенсионера, наличие иждивенцев, услуги телефонии, услуги интернета типа кабеля, сервиса OnlineSecurity и  техподдержка количество расторгнутых договоров пропорционально объему числа договоров в категории.


А вот в признаках наличие партнера, услуга OnlineBackup, DeviceProtection, StreamingTV, StreamingMovies и MultipleLines наоборот, чаще встречаются расторженные договоры среди абонентов использующих данную услугу.

## Обучение модели

Для обучения выбраны три модели классификации: случайный лес решающих деревьев, LightGBM и CatBoost. Для определения качества воспользуемся метрикой AUC ROC(площадь под кривой ошибок). Данная метрика показывает, на сколько верно обученная модель определяет пары классов. По заданному условию модель будет читаться корректной, если ROC_AUC > 0,85. Для большей точности результата воспользуемся методом обучения кросс-валидации. В данной работе коэффициент случайности примем за RANDOM_STATE = 150523.

Прежде чем перейти непосредственно к обучению разделим все данные на обучающие и тестовые. Размер тестовых данных выберем равным 25% от общего количества. Также преобразуем оставшиеся категориальные переменные с помощью OrdinalEncoder и выполним масштабирование числовых признаков с помощью StandardScaler().

### Деление данных на выборки

Преобразуем наши данные.

```python
RANDOM_STATE = 150523
```

```python
features_data = data_all.drop(['exited'], axis=1)
target_data = data_all['exited']
```

```python
features_train, features_test, target_train, target_test = (
            train_test_split(features_data,
                             target_data,
                             test_size=0.25,
                             random_state=RANDOM_STATE)
            )
```

```python
print('Размер обучающей выборки -',
      features_train.shape)
print('Размер тестовой выборки -',
      features_test.shape)
```

```python
cat_f = ['Type', 'PaymentMethod']
num_f = ['MonthlyCharges', 'duration']
```

```python
encoder = OrdinalEncoder(
    handle_unknown='use_encoded_value',
    unknown_value=-1
                        )
encoder.fit(features_train[cat_f])

features_train_ord = features_train.copy()
features_test_ord = features_test.copy()

features_train_ord[cat_f] = encoder.transform(features_train_ord[cat_f])
features_test_ord[cat_f] = encoder.transform(features_test_ord[cat_f])
```

```python
scaler = StandardScaler()

scaler.fit(features_train[num_f])

features_train_ord[num_f] = scaler.transform(features_train_ord[num_f])
features_test_ord[num_f] = scaler.transform(features_test_ord[num_f])
```

Проверим, как выполнилось преобразование.

```python
features_train_ord.sample(3)
```

```python
features_test_ord.sample(3)
```

Для упрощения обучения разных моделей напишем новую функцию. Для подбора лучших гиперпараметров моделей воспользуемся GridSearchCV. На вход функции будет подаваться обучаемая модель и перечень гиперпараметров для перебора. Разработанная функция будет возвращать лучшую обученную модель, результат ее метрики и время, затраченное на обучение модели.

```python
def model_scorer(model, param):

    start = time.time()

    grid_CV = GridSearchCV(estimator = model,
                           param_grid = param,
                           n_jobs=-1,
                           scoring='roc_auc',
                           refit = 'roc_auc',
                           cv=7)

    grid_CV.fit(features_train_ord, target_train)

    fit_time = time.time() - start

    model_gcv = grid_CV.best_estimator_
    auc_roc = grid_CV.best_score_

    print('Модель обучена. Лучшее значение'
           + ' по результатам кросс-валидации -',
          '%.3f' %auc_roc, '.')

    return model_gcv, auc_roc, fit_time
```

### Случайный лес решающих деревьев

Первой рассмотрим и обучим модель случайного леса решающих деревьев. Из возможных гиперпараметров для изменения выберем количество деревьев и глубину дерева.

```python
%%time
param_rfc = {'n_estimators' : range(100, 201, 10),
             'max_depth' : range(5, 10, 2)
            }
model_rfc, auc_roc_rfc, t_rfc = model_scorer(
                                     RandomForestClassifier(
                                        #class_weight='balanced',
                                        random_state=RANDOM_STATE
                                        ),
                                     param_rfc)
```

```python
%%time
param_lgbm = {'max_depth': range(2, 5, 2),
              'learning_rate': [0.1, 0.3],
              'num_leaves': range(12, 15, 2)
             }
model_lgbm, auc_roc_lgbm, t_lgbm = model_scorer(
                                     LGBMClassifier(
                                        boosting_type='gbdt',
                                        #class_weight='balanced',
                                        random_state=RANDOM_STATE
                                        ),
                                     param_lgbm)
```

### CatBoost

Последним рассмотрим модель градиентного бустинга CatBoost Classifier. Из возможных гиперпараметров для изменения выберем максимальную глубину деревьев и  максимальное число деревьев.

```python
%%time
cat_features = ['Type', 'PaymentMethod', 'Partner',
                'MultipleLines', 'OnlineBackup',
                'DeviceProtection']

param_cb = {'iterations': range(850, 951, 100),
            'depth': range(3, 7, 2),
            'learning_rate': [0.1, 0.15]
           }

model_cb, auc_roc_cb, t_cb = model_scorer(
                                     CatBoostClassifier(
                                        #loss_function='Logloss',
                                        silent=True,
                                        random_seed=RANDOM_STATE
                                        ),
                                     param_cb)
```

1. Столбцы с информацией о доте заключения и расторжения договора переведены в числовой формат
2. Для действующих договоров столбец с датой расторжения заполнили условно текущей датой(датой актуальности данных - 01.02.2020).

**3. Обучение модели.**

Для обучения выбраны три модели классификации: случайный лес решающих деревьев, LightGBM и CatBoost. Для определения качества воспользуемся метрикой AUC ROC(площадь под кривой ошибок). Данная метрика показывает, на сколько верно обученная модель определяет пары классов. По заданному условию модель будет читаться корректной, если ROC_AUC > 0,85. Для большей точности результата воспользуемся методом обучения кросс-валидации. В данной работе коэффициент случайности примем за RANDOM_STATE = 150523.
